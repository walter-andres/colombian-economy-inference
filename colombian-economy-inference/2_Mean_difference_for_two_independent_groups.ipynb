## **Mean difference for two independent groups**

Do men and women earn different average monthly labor incomes in Colombia?

Why this fits

* Two independent groups: Individuals are classified as either male or female, and each person belongs to only one group.
* Continuous outcome: Monthly labor income (COP) is a quantitative variable measured on a continuous scale.
* Large sample size: The Colombian household survey includes tens of thousands of employed individuals, providing high statistical power.
* Official open data: The data come from the DANE through the Gran Encuesta Integrada de Hogares (GEIH), which is publicly available at no cost.

Method:
 Independent samples t-test.

Hypotheses

Let:

* \mu_m = mean monthly labor income of men
* \mu_w = mean monthly labor income of women

The hypotheses are

* Null hypothesis (H₀):
    \mu_m=\mu_w
* Alternative hypothesis (H₁):
    \mu_m\neq\mu_w

This is a two-tailed independent samples t-test.

⸻

Assumptions

1. The two samples (men and women) are independent.
2. Monthly income is approximately normally distributed within each group (or the samples are sufficiently large).
3. Homogeneity of variances should be assessed using Levene’s Test.
    * If p>0.05, use the pooled t-test.
    * If p\le0.05, use Welch’s t-test.

⸻

Test statistic

If equal variances are assumed,

t=\frac{\bar X_m-\bar X_w}
{s_p\sqrt{\frac1{n_m}+\frac1{n_w}}}

where

s_p=\sqrt{\frac{(n_m-1)s_m^2+(n_w-1)s_w^2}
{n_m+n_w-2}}.

If variances are unequal, Welch’s t-test is preferred.

⸻

Decision rule

Using a significance level of

\alpha=0.05,

* If p-value < 0.05, reject H_0.
* If p-value ≥ 0.05, fail to reject H_0.

```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df = pd.read_csv(
    "/content/colombia_labor_market_by_sex.csv",
    sep=";",
    decimal=",",
    engine="python"
)
```

```python
# Rename first two columns
df = df.rename(columns={
    df.columns[0]: "Sex",
    df.columns[1]: "Concept"
})
```

```python
# ------------------------------------------------------------
# 2. Select occupation rate
# ------------------------------------------------------------
men = (
    df.loc[
        (df["Sex"].str.upper() == "HOMBRES") &
        (df["Concept"] == "Tasa de Ocupación (TO)"), # Specific concept for occupation rate
        df.columns[2:]
    ]
    .T.squeeze()
)

women = (
    df.loc[
        (df["Sex"].str.upper() == "MUJERES") &
        (df["Concept"] == "Tasa de Ocupación (TO)"), # Specific concept for occupation rate
        df.columns[2:]
    ]
    .T.squeeze()
)

# Convert to numeric
men = pd.to_numeric(men, errors="coerce").dropna()
women = pd.to_numeric(women, errors="coerce").dropna()
```

```python
# ------------------------------------------------------------
# 3. Hypotheses
# ------------------------------------------------------------

# H0: μ_m = μ_w
# H1: μ_m ≠ μ_w

alpha = 0.05
```

```python
# ------------------------------------------------------------
# 4. Descriptive statistics
# ------------------------------------------------------------

print("Descriptive Statistics")
print("-" * 40)
print(f"Men:")
print(f" Sample size = {len(men)}")
print(f" Mean = {men.mean():.2f}")
print(f" Std Dev = {men.std(ddof=1):.2f}\n")
print(f"Women:")
print(f" Sample size = {len(women)}")
print(f" Mean = {women.mean():.2f}")
print(f" Std Dev = {women.std(ddof=1):.2f}\n")
print
```
